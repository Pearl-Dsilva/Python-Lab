#### Visualize your dataset using Matplotlib and write a program to create line graph, bar graph, scatter plot and correlation graph or heat map.

```python
#scatter plot
import matplotlib.pyplot as plt
import pandas as pd

var = pd.read_excel("dataset.xlsx")
x=list(var['Animal'])
y=list(var['Weight(Kg)'])
plt.xticks(rotation = 50)
plt.xlabel('Animals')
plt.ylabel('Weights')
plt.title('Weight of different animals')
plt.scatter(x, y)
plt.show()
```

```python
# Correspondence of birth date and year of animals in zoo dataset
import matplotlib.pyplot as plt
import pandas as pd
from datetime import date

df = pd.read_excel("dataset.xlsx")
df['year'] = pd.DatetimeIndex(df['DateOfBirth']).year
today = date.today().year
df['age'] = today-df['year']

x = list(df['age'])
y = list(df['year'])
plt.plot(x, y)
plt.show()
```

```python
# heatmap showing corelation of weight and health status of animals in zoo dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel("dataset.xlsx")

df['HealthStatus'] = df['HealthStatus'].map({'Healthy': 1, 'Unhealthy': 0})

subset_df = df[['Weight(Kg)', 'HealthStatus']]

correlation_matrix = subset_df.corr()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, annot_kws={"size": 12})
plt.title('Correlation Heatmap: Weight(Kg) vs. HealthStatus')
plt.show()
```
